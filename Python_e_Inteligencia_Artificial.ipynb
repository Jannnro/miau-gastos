Exemplo2: Calculando as medias do aluno.

```python
media = float(input("Digite a sua média: "))

if media >= 7.0:
    print("Você passou de ano!")
elif media >= 5.0:
    print("Você está de recuperação")
else:
  print("Você foi reprovado")
```

Se fizermos ao contrario, ou seja

```python
media = float(input("Digite a sua média: "))

if media < 7.0:
    print("Você passou de ano!")
elif media >= 5.0:
    print("Você está de recuperação")
else:
  print("Você foi reprovado")
```

e a sua nota for maior que 5, voce estará de recuperação de qualquer maneira pois a condição do igual ou maior que 5 já foi atendida.
para resovermos isso, vamos colocar um novo operador junto com elif que é o AND

Nesse caso, o elif and viram duas operações em uma unica onde, se a media for igual ou maior que 5 e menor do que 7, voce está de recuperação
qualquer valor acima de 7 já está aprovado

```python
media = float(input("Digite a sua média: "))

if media < 5.0:
    print("Você foi reprovado!")
elif media >= 5.0 and media < 7.0:
    print("Você está de recuperação")
else:
  print("Você foi aprovado")
```

Nesse caso, estamos fazendo uma lista de medias para cada aluno e somando +1 ponto nas medias
Entao nós definimos a variavel N recebendo a posição 0 (0, 1, 2, 3, 4)
  Enquanto N for MENOR que 5, ele adicionaria +1 na media para todas as posições.
Abaixo (no IF) no definimos que: SE o valor de alguma determinada posição fosse MAIOR que 10, ele receberia o valor 10 para aquela posição
  no N+N+1 estamos atribuindo +1 para todas as posições que forem menor que 10

```python
n = 0

while n < 5:
  listaMedias[n] = (listaMedias[n] + 1.0)
  if (listaMedias[n] > 10.0):
      listaMedias[n] = 10.0
  n = n + 1


print(listaMedias)
```
